```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('salaries.csv')
df.head()
```

```python
input=df.drop('salary_more_then_100k',axis='columns')
target=df['salary_more_then_100k']
```

```python
input
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
company_le=LabelEncoder()
job_le=LabelEncoder()
degree_le=LabelEncoder()
```

```python
input['company_n']=company_le.fit_transform(input['company'])
input['job_n']=job_le.fit_transform(input['job'])
input['degree_n']=degree_le.fit_transform(input['degree'])
input.head()
```

```python
input_n=input.drop(['company','job','degree'],axis='columns')
input_n
```

```python
from sklearn import tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(input_n,target)
```

```python
model.score(input_n,target)
```

```python
model.predict([[2,1,1]])
```
