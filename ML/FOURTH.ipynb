```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('town_areavsprice.csv')
df
```

```python
m_r_w=pd.get_dummies(df.town)
m_r_w.head()
```

```python
new_df=pd.concat([df,m_r_w],axis='columns')
new_df
```

```python
final=new_df.drop(['town','west'],axis='columns')
final
```

```python
model=LinearRegression()
```

```python
x=final.drop(['price'],axis='columns')
x
```

```python
y=final['price']
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(x,y)
```
